1. **Veri Görselleştirme Nedir?**

Verileri görselleştirme, karmaşık ve dağınık verileri düzenleyerek kolay anlaşılır, yorumlanabilir hale getirmektir. Bir başka ifade ile veriyi görselleştirme, gördüğü ki soyut verilerin görsel hale getirilerek somutlaştırılması ve bundan bir ön bilginin elde edilmesi işidir.
## Kategorik değişken: sütun grafik. countplot bar
## Sayısal değişken: hist, boxplot

**Kategorik Değişken Görselleştirme**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 500)
df = sns.load_dataset("titanic")
df.head()
```

```python
df['sex'].value_counts().plot(kind='bar')
plt.show()
```

**Sayısal Değişken Görselleştirme**

```python
plt.hist(df["age"])
plt.show()
```

```python
plt.boxplot(df["fare"])
plt.show()
```

**Matplotlib'in Özellikleri**

Matplotlib; verileştirmede temel python kütüphanesidir. 2 ve 3 boyutlu çizimler yapmasını sağlar.

**plot**

değişkenlerin görsele aktarımını sağlar

```python
import numpy as np
x = np.array([2, 4, 6, 8, 10])
y = np.array([1, 3, 5, 7, 9])

plt.plot(x, y)
plt.show()
```

**marker**

veriler üzerinde işaretleme yapmamızı sağlar

```python
y = np.array([13, 28, 11, 100])

plt.plot(y, marker='o')
plt.show()
```

**line**

çizgi grafiklerinde kullanılır

```python
y = np.array([13, 28, 11, 100])
plt.plot(y, linestyle="dashdot", color="r")
plt.show()
```

**Multiple Lines**

bir grafik üzerinde iki veriye ulaşımı olanak sağlar.

```python
x = np.array([23, 18, 31, 10])
y = np.array([13, 28, 11, 100])
plt.plot(x)
plt.plot(y)
plt.show()
```

**Labels**

grafikler üzerindeki eksenlere ve grafiğe isim vermeyi sağlar

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.plot(x, y)
# Başlık
plt.title("Bu ana başlık")

# X eksenini isimlendirme
plt.xlabel("X ekseni isimlendirmesi")

plt.ylabel("Y ekseni isimlendirmesi")

plt.grid()
plt.show()
```

#**SEABORN**

**Seaborn**, MatPlotLib kütüphanesini temel alan bir Python görselleştirme kütüphanesidir. Pandas kütüphanesinin veri yapısına da kolayca entegre edilebilir. Özellikle istatistiksel grafikler için çok kullanışlı bir kod yapısı sağlar.

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
df = sns.load_dataset("tips")
df.head()

df["sex"].value_counts()
sns.countplot(x=df["sex"], data=df)
plt.show()

df['sex'].value_counts().plot(kind='bar')
plt.show()
```

**Sayısal Değişken Görselleştirme**

```python
sns.boxplot(x=df["total_bill"])
plt.show()

df["total_bill"].hist()
plt.show()
```
