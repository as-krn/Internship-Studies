**Değişkenler Üzerinde İşlemler**

```python
col_names = ["Cost of Revenue", "Gross Profit", "Premium Revenue"]
df[col_names].head()
```

```python
df["Cost of Revenue2"] = df["Cost of Revenue"]**2
df["Cost of Revenue3"] = df["Cost of Revenue"] / df["Cost of Revenue2"]# yaptığımız matematiksel işlemler son sütunlara eklenir.
df.head()
```

**iloc & loc**


1.   iloc: tamsayı tabanlı seçim
2.   loc: etiket tabanlı seçim




```python
df.iloc[0:3]# ilk 3 satırı getirir
```

```python
df.iloc[0:3, 0:3]#ilk 3 satır ve ilk 3 sütunu getirir
```

```python
df.loc[0:3, "Gross Profit"]#girilen sütuna göre satır getirir
```

```python
col_names = ["Cost of Revenue", "Gross Profit", "Premium Revenue"]
df.loc[0:3, col_names]
```

**Toplulaştırma ve Gruplama (Aggregation & Grouping)**

```python
df["Gross Profit"].mean()#ortalama almaya yarar
```

```python
df.groupby("Gross Profit")["Premium Revenue"].mean()#gruplamak için groupby() kullanılır
```

```python
df.groupby("Gross Profit").agg({"Premium Revenue": ["mean", "sum"]})# sözlük yapısı ile ortalama , toplam ve min max  işlemleri yapılabilir
```

**Pivot table**

PivotTable, verilerinize ilişkin karşılaştırmaları, kalıpları ve eğilimleri görmek amacıyla verileri hesaplamanıza, özetlemenize ve analiz etmenize yönelik güçlü bir araçtır.

```python
df.pivot_table("Cost of Revenue", ["Premium Revenue", "Premium MAUs"])
```

**Apply ve Lambda**

apply() içerisinde kullan at fonksiyonu olan lambda kullanarak ek fonksiyon kullanmaya gerek kalmaz.

```python
df.loc[:, df.columns.str.contains("Premium MAUs")].apply(lambda x: x/10).head()
```

**Birleştirme (Join) İşlemleri**

1.   Merge ile Birleştirme İşlemleri
2.   Concat ile birleştirme işlemleri



```python
df1 = pd.DataFrame({'employees': ['john', 'dennis', 'mark', 'maria'],
                    'group': ['accounting', 'engineering', 'engineering', 'hr']})

df2 = pd.DataFrame({'employees': ['mark', 'john', 'dennis', 'maria'],
                    'start_date': [2010, 2009, 2014, 2019]})

pd.merge(df1, df2)
pd.merge(df1, df2, on="employees")
```

```python
import numpy as np
m = np.random.randint(1, 30, size=(5, 3))
df1 = pd.DataFrame(m, columns=["var1", "var2", "var3"])
df2 = df1 + 99

pd.concat([df1, df2], ignore_index=True)
```
